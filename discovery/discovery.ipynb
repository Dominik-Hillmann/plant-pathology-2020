

# Plant Pathology 2020 Challenge: Data Discovery

In this challenge, the goal is to classify images of leaves into healthy ones and ones that have a one or more diseases.

## Structure of Raw Data

```python
# Python libraries
import os

# External modules
import pandas as pd


raw_data_path = os.path.join('..', 'data', 'raw-data')
print(os.listdir(raw_data_path))
print()

train_data = pd.read_csv(os.path.join(raw_data_path, 'train.csv'))
test_data = pd.read_csv(os.path.join(raw_data_path, 'test.csv'))
submission = pd.read_csv(os.path.join(raw_data_path, 'sample_submission.csv'))

print('Train data sample:', train_data.shape)
display(train_data.sample(15))
print('Test data sample:', test_data.shape)
display(test_data.sample(15))
print('Sample submission:', submission.shape)
display(submission.sample(15))
```

## A look at a few examples

```python
# Python libraries
from random import choice

# External modules
import cv2
import matplotlib.pyplot as plt


imgs = os.listdir(os.path.join(raw_data_path, 'images'))
imgs = [ img for img in imgs if 'Train' in img ]
sample_imgs = []
for _ in range(5):
    sample_imgs.append(choice(imgs))

loaded_imgs = []
for img_name in sample_imgs:
    img = cv2.imread(os.path.join(raw_data_path, 'images', img_name))
    loaded_imgs.append(img)
    
sample_imgs = [ img_name[:-4] for img_name in sample_imgs ]
print(sample_imgs)
```

## Distribution of labels

```python
healthy = train_data[train_data['healthy'] == 1]
print('Sample of healthy leaves:', healthy.shape)
display(healthy.sample(10))
```

```python
mutliple_diseases = train_data[train_data['multiple_diseases'] == 1]
print('Sample of multiply diseased leaves:', mutliple_diseases.shape)
display(mutliple_diseases.sample(10))
```

```python
rust = train_data[train_data['rust'] == 1]
print('Sample of leaves with rust:', rust.shape)
display(rust.sample(10))
```

```python
scab = train_data[train_data['scab'] == 1]
print('Sample of leaves with scab:', scab.shape)
display(scab.sample(10))
```

```python
total = train_data.shape[0]
summed = len(healthy) + len(mutliple_diseases) + len(rust) + len(scab)

plt.pie([
    len(healthy) / total,
    len(mutliple_diseases) / total,
    len(rust) / total,
    len(scab) / total
], 
    colors = ['green', 'red', 'brown', 'orange'],
    labels = ['Healthy', 'Multiple diseases', 'Rust', 'Scab']
)

print(summed, total, summed == total)
print('All labels mutually exclude each other, otherwise the sum of all labels would be larger.')
```
