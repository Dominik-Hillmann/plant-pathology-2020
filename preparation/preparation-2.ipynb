```python
# Python libraries
import os
import random

# External modules
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Constants
SEED = 69

# Settings
random.seed(SEED)


raw_data_path = os.path.join('..', 'data', 'raw-data')
img_names = os.listdir(os.path.join(raw_data_path, 'images'))

img_sample = []
for _ in range(10):
    rand_name = random.choice(img_names)
    rand_img = cv2.imread(os.path.join(raw_data_path, 'images', rand_name))
    img_sample.append(rand_img)
```

```python
test_img = img_sample[0]
plt.imshow(test_img)
```
